```python
URL = 'https://flowersanalytics.com/projects/wdi-income-inequality/income_inequality.html'
```

```python
import qrcode
from IPython.display import display, Image

# Create QR code instance
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(URL)
qr.make(fit=True)

# Create image from the QR Code
qr_img = qr.make_image(fill_color="black", back_color="white")

# Display the QR code in the notebook
display(qr_img)

# Optionally save the QR code
qr_img.save("url_qrcode.png")
```
